### 二、装饰器（Decorator）
装饰器本质是包裹函数的函数，让你在不改原函数的情况下添加功能。AI 项目里最常用于日志、重试、缓存。

```python
import time
import functools

# 最简单的装饰器：计时器
def timer(func):
    @functools.wraps(func)  # 保留原函数的名字和文档
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        elapsed = time.time() - start
        print(f"{func.__name__} 耗时 {elapsed:.2f}s")
        return result
    return wrapper

@timer
def call_llm(prompt: str) -> str:
    time.sleep(1)  # 模拟 API 调用
    return "模型回复"

call_llm("你好")  # 输出：call_llm 耗时 1.00s
```

带参数的装饰器（更实用）：

```python
# 自动重试装饰器 —— API 调用超时时非常有用
def retry(max_attempts: int = 3, delay: float = 1.0):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            last_error = None
            for attempt in range(1, max_attempts + 1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    last_error = e
                    print(f"第 {attempt} 次失败：{e}，{'重试中...' if attempt < max_attempts else '放弃'}")
                    if attempt < max_attempts:
                        time.sleep(delay)
            raise last_error
        return wrapper
    return decorator

@retry(max_attempts=3, delay=2.0)
def unstable_api_call(prompt: str) -> str:
    # 模拟偶发性失败
    import random
    if random.random() < 0.6:
        raise ConnectionError("API 超时")
    return "成功响应"

# unstable_api_call("Nihao")
```

### 三、异步编程（Asyncio）
这是 AI 项目里最重要的一块。调用 LLM API 时需要等待网络响应，如果用同步代码，等待期间 CPU 完全空闲。用异步就能同时发出多个请求。

先理解核心概念：

```python
import asyncio
import nest_asyncio

# async def 定义协程函数
async def fetch_response(prompt: str) -> str:
    print(f"开始处理：{prompt}")
    await asyncio.sleep(2)  # await = 挂起，让出控制权给其他协程
    print(f"完成：{prompt}")
    return f"回复：{prompt}"

# 运行单个协程
async def main():
    result = await fetch_response("你好")
    print(result)

asyncio.run(main())
```

关键：并发执行多个请求

```python
import asyncio
import time

async def call_llm(prompt: str, delay: float) -> str:
    """模拟 LLM API 调用"""
    await asyncio.sleep(delay)
    return f"[{prompt}] 的回复"

async def main():
    prompts = ["问题1", "问题2", "问题3", "问题4"]
    
    # ❌ 顺序执行：总耗时 = 所有请求时间之和
    start = time.time()
    results_seq = []
    for p in prompts:
        r = await call_llm(p, delay=1.0)
        results_seq.append(r)
    print(f"顺序执行耗时：{time.time()-start:.1f}s")  # ~4.0s
    
    # ✅ 并发执行：总耗时 ≈ 最慢那个请求的时间
    start = time.time()
    tasks = [call_llm(p, delay=1.0) for p in prompts]
    results_con = await asyncio.gather(*tasks)
    print(f"并发执行耗时：{time.time()-start:.1f}s")  # ~1.0s
    
    print(results_con)

asyncio.run(main())
```

实战：用 httpx 异步调用 API

```python
import asyncio
import httpx
from typing import List

async def fetch_one(
    client: httpx.AsyncClient,
    prompt: str,
    semaphore: asyncio.Semaphore,
) -> str:
    """用信号量限制并发数，避免触发 API 限速"""
    async with semaphore:
        response = await client.post(
            "https://api.example.com/chat",
            json={"prompt": prompt},
            timeout=30.0,
        )
        response.raise_for_status()
        return response.json()["content"]

async def batch_call(prompts: List[str], max_concurrent: int = 5) -> List[str]:
    semaphore = asyncio.Semaphore(max_concurrent)  # 最多同时 5 个请求
    
    async with httpx.AsyncClient() as client:
        tasks = [fetch_one(client, p, semaphore) for p in prompts]
        results = await asyncio.gather(*tasks, return_exceptions=True)
    
    # 处理部分失败的情况
    final = []
    for r in results:
        if isinstance(r, Exception):
            print(f"请求失败：{r}")
            final.append(None)
        else:
            final.append(r)
    return final

asyncio.run(batch_call(["提问1", "提问2", ...]))
```

### 四、API 调用（Anthropic SDK）
把前三块结合起来，这是实际项目中的写法：

```python
import asyncio
from typing import List, Optional
from dataclasses import dataclass, field
import anthropic
import os

os.environ["ANTHROPIC_API_KEY"] = "YOUR_ANTHROPIC_API_KEY"  # 替换成你的真实 key

# 类型注解定义结构
@dataclass
class ConversationMessage:
    role: str
    content: str

@dataclass
class Conversation:
    messages: List[ConversationMessage] = field(default_factory=list)
    system: str = "你是一个 AI Agent 学习助手"
    
    def add(self, role: str, content: str) -> None:
        self.messages.append(ConversationMessage(role, content))
    
    def to_api_format(self) -> List[dict]:
        return [{"role": m.role, "content": m.content} for m in self.messages]


# 装饰器：自动重试
def retry(max_attempts: int = 3):
    def decorator(func):
        import functools
        @functools.wraps(func)
        async def wrapper(*args, **kwargs):
            for attempt in range(max_attempts):
                try:
                    return await func(*args, **kwargs)
                except anthropic.RateLimitError:
                    wait = 2 ** attempt  # 指数退避：1s, 2s, 4s
                    print(f"触发限速，等待 {wait}s 后重试...")
                    await asyncio.sleep(wait)
            raise RuntimeError("超过最大重试次数")
        return wrapper
    return decorator


# 异步 API 调用
@retry(max_attempts=3)
async def chat(
    conversation: Conversation,
    model: str = "claude-sonnet-4-6",
    max_tokens: int = 1024,
) -> str:
    client = anthropic.AsyncAnthropic()  # 异步客户端
    
    response = await client.messages.create(
        model=model,
        max_tokens=max_tokens,
        system=conversation.system,
        messages=conversation.to_api_format(),
    )
    return response.content[0].text


# 流式输出（打字机效果）
async def chat_stream(prompt: str) -> None:
    client = anthropic.AsyncAnthropic()
    
    async with client.messages.stream(
        model="claude-sonnet-4-6",
        max_tokens=1024,
        messages=[{"role": "user", "content": prompt}],
    ) as stream:
        async for text in stream.text_stream:
            print(text, end="", flush=True)
    print()  # 换行


# 完整使用示例
async def main():
    conv = Conversation(system="你是一个简洁的助手，回答不超过50字")
    
    # 多轮对话
    questions = ["什么是 RAG？", "它和微调有什么区别？"]
    
    for q in questions:
        conv.add("user", q)
        reply = await chat(conv)
        conv.add("assistant", reply)
        print(f"Q: {q}")
        print(f"A: {reply}\n")
    
    # 流式输出
    print("流式输出示例：")
    await chat_stream("用一句话解释 Transformer")

asyncio.run(main())
```

### 二、路由（Routes）
路由就是"什么 URL 对应什么函数"，FastAPI 用类型注解自动做参数校验。

```python
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
from typing import Optional

app = FastAPI()

# ── 请求/响应结构用 Pydantic 定义 ──────────────────────────
class ChatRequest(BaseModel):
    message: str
    session_id: Optional[str] = None
    max_tokens: int = 1024

class ChatResponse(BaseModel):
    reply: str
    session_id: str
    tokens_used: int

# ── GET：查询参数自动从 URL 读取 ────────────────────────────
@app.get("/health")
async def health_check():
    return {"status": "ok"}

@app.get("/models")
async def list_models(provider: str = "anthropic"):  # ?provider=anthropic
    models = {
        "anthropic": ["claude-sonnet-4-6", "claude-opus-4-6"],
        "openai": ["gpt-4o", "gpt-4o-mini"],
    }
    if provider not in models:
        raise HTTPException(status_code=404, detail=f"Provider '{provider}' not found")
    return {"provider": provider, "models": models[provider]}

# ── POST：请求体自动从 JSON 解析 ────────────────────────────
@app.post("/chat", response_model=ChatResponse)
async def chat(request: ChatRequest):
    # 实际项目里这里调用 LLM
    reply = f"收到你的消息：{request.message}"
    return ChatResponse(
        reply=reply,
        session_id=request.session_id or "default",
        tokens_used=len(request.message),
    )

# ── 路径参数 ────────────────────────────────────────────────
@app.get("/sessions/{session_id}")
async def get_session(session_id: str):
    return {"session_id": session_id, "messages": []}

@app.delete("/sessions/{session_id}")
async def delete_session(session_id: str):
    return {"deleted": session_id}
```

### 三、中间件（Middleware）
中间件是每个请求都会经过的"拦截层"，用于日志、鉴权、限速等横切逻辑。

```python
import time
import uuid
from fastapi import FastAPI, Request
from fastapi.middleware.cors import CORSMiddleware

app = FastAPI()

# ── 1. CORS（前端跨域必须加）──────────────────────────────
app.add_middleware(
    CORSMiddleware,
    allow_origins=["http://localhost:3000"],  # 前端地址，生产环境填真实域名
    allow_methods=["*"],
    allow_headers=["*"],
)

# ── 2. 自定义中间件：请求日志 + 计时 ──────────────────────
@app.middleware("http")
async def log_requests(request: Request, call_next):
    request_id = str(uuid.uuid4())[:8]
    start = time.time()

    print(f"[{request_id}] → {request.method} {request.url.path}")

    response = await call_next(request)  # 执行实际的路由函数

    elapsed = (time.time() - start) * 1000
    print(f"[{request_id}] ← {response.status_code} ({elapsed:.1f}ms)")

    response.headers["X-Request-ID"] = request_id
    return response

# ── 3. API Key 鉴权中间件 ──────────────────────────────────
VALID_API_KEYS = {"test-key-123", "prod-key-456"}

@app.middleware("http")
async def auth_middleware(request: Request, call_next):
    # 放行不需要鉴权的路由
    if request.url.path in ["/", "/health", "/docs", "/openapi.json"]:
        return await call_next(request)

    api_key = request.headers.get("X-API-Key")
    if not api_key or api_key not in VALID_API_KEYS:
        from fastapi.responses import JSONResponse
        return JSONResponse(
            status_code=401,
            content={"detail": "Invalid or missing API key"},
        )
    return await call_next(request)
```

### 四、把 LLM 接进来（完整 AI 后端）
把前面学的 async + Anthropic SDK 接到 FastAPI 路由里：

```python
import asyncio
import anthropic
from fastapi import FastAPI, HTTPException
from fastapi.responses import StreamingResponse
from pydantic import BaseModel
from typing import Optional
import json

app = FastAPI(title="AI Agent Backend")
client = anthropic.AsyncAnthropic()  # 全局复用，不要每次请求都新建

# 简单的内存会话存储（生产环境换 Redis）
sessions: dict[str, list] = {}

class ChatRequest(BaseModel):
    message: str
    session_id: str = "default"
    system: str = "你是一个简洁的 AI 助手，回答不超过100字"

# ── 普通对话接口 ───────────────────────────────────────────
@app.post("/chat")
async def chat(req: ChatRequest):
    # 维护多轮历史
    if req.session_id not in sessions:
        sessions[req.session_id] = []

    sessions[req.session_id].append({"role": "user", "content": req.message})

    try:
        response = await client.messages.create(
            model="claude-sonnet-4-6",
            max_tokens=1024,
            system=req.system,
            messages=sessions[req.session_id],
        )
        reply = response.content[0].text
        sessions[req.session_id].append({"role": "assistant", "content": reply})
        return {"reply": reply, "session_id": req.session_id}

    except anthropic.BadRequestError as e:
        raise HTTPException(status_code=400, detail=str(e))

# ── 流式对话接口（打字机效果）─────────────────────────────
@app.post("/chat/stream")
async def chat_stream(req: ChatRequest):
    async def generate():
        async with client.messages.stream(
            model="claude-sonnet-4-6",
            max_tokens=1024,
            system=req.system,
            messages=[{"role": "user", "content": req.message}],
        ) as stream:
            async for text in stream.text_stream:
                # SSE 格式：前端用 EventSource 接收
                yield f"data: {json.dumps({'text': text})}\n\n"
        yield "data: [DONE]\n\n"

    return StreamingResponse(generate(), media_type="text/event-stream")

# ── 清除会话 ───────────────────────────────────────────────
@app.delete("/sessions/{session_id}")
async def clear_session(session_id: str):
    sessions.pop(session_id, None)
    return {"cleared": session_id}
```

```python
# ============================================================
# 安装依赖（在终端运行，不是在 Python 里运行）
# pip install fastapi uvicorn anthropic pydantic
# 启动命令：uvicorn main:app --reload --port 8000
# 查看自动生成的 API 文档：http://127.0.0.1:8000/docs
# ============================================================

import asyncio
import time
import uuid
import json
import anthropic

from fastapi import FastAPI, HTTPException, Request
from fastapi.middleware.cors import CORSMiddleware
from fastapi.responses import StreamingResponse, JSONResponse
from pydantic import BaseModel
from typing import Optional


# ============================================================
# 第一部分：创建 FastAPI 应用实例
# ============================================================

# FastAPI() 创建整个应用，title 和 version 会显示在自动文档里
app = FastAPI(
    title="AI Agent Backend",
    version="1.0",
    description="AI 对话后端，支持多轮对话和流式输出",
)

# 创建 Anthropic 异步客户端
# 注意：放在模块级别全局复用，避免每次请求都新建连接（浪费资源）
# 它会自动从环境变量 ANTHROPIC_API_KEY 读取 key
client = anthropic.AsyncAnthropic()

# 简单的内存字典模拟会话存储
# key = session_id（字符串），value = 该会话的历史消息列表
# ⚠️ 仅用于学习演示，进程重启后数据清空
# 生产环境应换成 Redis 或数据库
sessions: dict[str, list] = {}


# ============================================================
# 第二部分：数据模型（用 Pydantic 定义请求/响应结构）
# ============================================================

# Pydantic 的 BaseModel 会自动：
#   1. 解析 JSON 请求体，转换为 Python 对象
#   2. 校验字段类型（类型不符会自动返回 422 错误）
#   3. 生成 API 文档里的参数说明

class ChatRequest(BaseModel):
    message: str                              # 必填，用户发送的消息
    session_id: str = "default"              # 选填，会话 ID，用于多轮对话
    system: str = "你是一个简洁的 AI 助手，回答不超过100字"  # 选填，系统提示词


class ChatResponse(BaseModel):
    reply: str          # 模型的回复内容
    session_id: str     # 回传 session_id，方便前端继续追踪


# ============================================================
# 第三部分：中间件（每个请求都会经过，按注册顺序执行）
# ============================================================

# ── 中间件 1：CORS（跨域资源共享）──────────────────────────
# 前端（如 React/Vue）和后端跑在不同端口时，浏览器会拦截请求
# 加这个中间件告诉浏览器"允许跨域"
app.add_middleware(
    CORSMiddleware,
    allow_origins=["http://localhost:3000"],  # 允许的前端地址
                                              # 生产环境改为真实域名，不要用 "*"
    allow_methods=["*"],    # 允许所有 HTTP 方法（GET/POST/DELETE 等）
    allow_headers=["*"],    # 允许所有请求头
)

# ── 中间件 2：请求日志 + 耗时统计 ──────────────────────────
# @app.middleware("http") 表示这是一个 HTTP 中间件
# 每个请求进来 → 先执行这里 → 再执行路由函数 → 再回到这里 → 返回响应
@app.middleware("http")
async def log_requests(request: Request, call_next):
    # 生成一个短 ID，用于在日志里追踪同一次请求的进出
    request_id = str(uuid.uuid4())[:8]
    start = time.time()

    # 记录请求进入
    print(f"[{request_id}] → {request.method} {request.url.path}")

    # call_next(request) = 把请求交给下一个处理者（路由函数）
    # 这里用 await 等待路由函数执行完毕
    response = await call_next(request)

    # 计算耗时（毫秒）
    elapsed = (time.time() - start) * 1000
    print(f"[{request_id}] ← {response.status_code} ({elapsed:.1f}ms)")

    # 在响应头里附上请求 ID，方便前端调试
    response.headers["X-Request-ID"] = request_id
    return response

# ── 中间件 3：API Key 鉴权 ──────────────────────────────────
# 所有请求必须在 Header 里带上 X-API-Key，否则拒绝访问
VALID_API_KEYS = {"test-key-123", "prod-key-456"}  # 实际项目从数据库或环境变量读取

@app.middleware("http")
async def auth_middleware(request: Request, call_next):
    # 白名单：这些路径不需要鉴权
    # /docs 和 /openapi.json 是 FastAPI 自动生成的文档页面
    PUBLIC_PATHS = ["/", "/health", "/docs", "/openapi.json", "/redoc"]
    if request.url.path in PUBLIC_PATHS:
        return await call_next(request)   # 直接放行

    # 从请求头读取 X-API-Key
    api_key = request.headers.get("X-API-Key")

    # key 不存在或不合法 → 返回 401 未授权
    if not api_key or api_key not in VALID_API_KEYS:
        return JSONResponse(
            status_code=401,
            content={"detail": "Invalid or missing API key. Add 'X-API-Key' header."},
        )

    # key 合法 → 放行，继续执行路由函数
    return await call_next(request)


# ============================================================
# 第四部分：路由（定义 URL 和对应的处理函数）
# ============================================================

# ── GET /health：健康检查接口 ──────────────────────────────
# 通常用于 Docker / K8s 检测服务是否存活
@app.get("/health")
async def health_check():
    return {"status": "ok"}


# ── GET /：根路径欢迎信息 ──────────────────────────────────
@app.get("/")
async def root():
    return {"message": "AI backend is running. Visit /docs for API documentation."}


# ── POST /chat：普通对话接口 ────────────────────────────────
# response_model=ChatResponse 告诉 FastAPI：
#   1. 用 ChatResponse 验证返回值
#   2. 在文档里展示响应结构
@app.post("/chat", response_model=ChatResponse)
async def chat(req: ChatRequest):
    """
    多轮对话接口。
    - 相同 session_id 的请求共享历史消息
    - 模型会记住同一 session 内的上下文
    """
    # 如果这是该 session 的第一条消息，初始化空列表
    if req.session_id not in sessions:
        sessions[req.session_id] = []

    # 把用户消息追加到会话历史
    # Anthropic API 要求消息格式：{"role": "user"/"assistant", "content": "..."}
    sessions[req.session_id].append({
        "role": "user",
        "content": req.message,
    })

    try:
        # 调用 Anthropic API
        # 传入完整历史消息，模型才能理解上下文
        response = await client.messages.create(
            model="claude-sonnet-4-6",
            max_tokens=1024,
            system=req.system,                    # 系统提示词，定义模型角色
            messages=sessions[req.session_id],    # 完整历史（含本次用户消息）
        )

        # 从响应中提取文本内容
        # response.content 是列表，[0] 是第一个内容块，.text 是文本
        reply = response.content[0].text

        # 把模型回复也存入历史，供下一轮对话使用
        sessions[req.session_id].append({
            "role": "assistant",
            "content": reply,
        })

        return ChatResponse(reply=reply, session_id=req.session_id)

    except anthropic.AuthenticationError:
        # API Key 无效
        raise HTTPException(status_code=401, detail="Anthropic API key is invalid.")

    except anthropic.BadRequestError as e:
        # 请求参数有问题（如余额不足、模型名称错误等）
        raise HTTPException(status_code=400, detail=str(e))

    except anthropic.RateLimitError:
        # 触发限速，让客户端稍后重试
        raise HTTPException(status_code=429, detail="Rate limit exceeded. Please retry later.")


# ── POST /chat/stream：流式对话接口 ─────────────────────────
# 使用 SSE（Server-Sent Events）协议，边生成边发送
# 前端用 EventSource 或 fetch + ReadableStream 接收
@app.post("/chat/stream")
async def chat_stream(req: ChatRequest):
    """
    流式对话接口（打字机效果）。
    返回 text/event-stream 格式，每个 token 生成后立即推送给客户端。
    """

    # 定义一个异步生成器函数，逐块 yield 数据
    async def generate():
        async with client.messages.stream(
            model="claude-sonnet-4-6",
            max_tokens=1024,
            system=req.system,
            messages=[{"role": "user", "content": req.message}],
        ) as stream:
            # stream.text_stream 是异步迭代器，每次 yield 一小块文本
            async for text in stream.text_stream:
                # SSE 协议格式：必须以 "data: " 开头，以 "\n\n" 结尾
                # json.dumps 确保特殊字符（引号、换行）被正确转义
                yield f"data: {json.dumps({'text': text})}\n\n"

        # 发送结束标志，前端收到后停止监听
        yield "data: [DONE]\n\n"

    # StreamingResponse 会持续把 generate() yield 出来的内容推给客户端
    # media_type="text/event-stream" 是 SSE 协议规定的 Content-Type
    return StreamingResponse(generate(), media_type="text/event-stream")


# ── DELETE /sessions/{session_id}：清除会话历史 ─────────────
# {session_id} 是路径参数，FastAPI 会自动提取并传入函数
@app.delete("/sessions/{session_id}")
async def clear_session(session_id: str):
    """清除指定会话的历史消息，下次对话将从头开始。"""
    # dict.pop(key, None)：key 存在则删除并返回值，不存在返回 None（不报错）
    removed = sessions.pop(session_id, None)
    if removed is None:
        raise HTTPException(status_code=404, detail=f"Session '{session_id}' not found.")
    return {"cleared": session_id, "messages_removed": len(removed)}


# ── GET /sessions/{session_id}：查看会话历史 ───────────────
@app.get("/sessions/{session_id}")
async def get_session(session_id: str):
    """返回指定会话的完整历史消息。"""
    if session_id not in sessions:
        raise HTTPException(status_code=404, detail=f"Session '{session_id}' not found.")
    return {
        "session_id": session_id,
        "message_count": len(sessions[session_id]),
        "messages": sessions[session_id],
    }


# ============================================================
# 第五部分：本地运行入口（直接 python main.py 时使用）
# ============================================================

# 这段代码只在直接运行 python main.py 时执行
# 用 uvicorn 命令启动时不会执行（uvicorn 自己管理启动）
if __name__ == "__main__":
    import uvicorn
    # host="0.0.0.0" 让局域网内其他设备也能访问
    # reload=True 代码修改后自动重启，仅用于开发环境
    uvicorn.run("main:app", host="0.0.0.0", port=8000, reload=True)
```
